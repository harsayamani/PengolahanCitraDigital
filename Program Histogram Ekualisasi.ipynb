```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('download.jpg', 0)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
hist = cv2.calcHist([img], channels=[0], mask=None, histSize=[256], ranges=[0, 256])
plt.imshow(img)
```

```python
#display normalisasi
norm = hist.cumsum()
normalized = norm + hist.max() / norm.max()
plt.plot(normalized, color='b')
```

```python
#display histogram
plt.hist(img.flatten(), 256, [0,256], color='r')
plt.xlim([0,256])
plt.legend(('norm', 'histogram'), loc='upper left')
plt.show()
```

```python
#histogram equalisasi
grayimg = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
histeq = cv2.equalizeHist(grayimg)
```

```python
cv2.imshow('original', img)
cv2.imshow('Hasil Histogram Ekualisasi', histeq)
cv2.waitKey(0)
```
