```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def opr_negative(img):
    d = img.shape
    row = d[0]
    coloumn = d[1]
    layer = d[2]
    
    out = img
    
    for x in range (0, row):
        for y in range (0, coloumn):
            for l in range (0, layer):
                out[x, y, l] = 255 - img[x, y, l]
                
#     cv2.namedWindow('gambar negatif', cv2.WINDOW_AUTOSIZE)
#     cv2.imshow('gambar negatif', out)
    
#     tombol = cv2.waitKey(0) & 0xFF
#     if tombol==ord('s'):
#         cv2.imwrite('gambarNegatif.png', img)
    
    return out

if __name__ == "__main__":
    pict = cv2.imread('download.jpg')
    pict = cv2.cvtColor(pict, cv2.COLOR_BGR2RGB)
    plt.subplot(1,2,1)
    plt.imshow(pict)
    plt.subplot(1,2,2)
    plt.imshow(opr_negative(pict))
#     plt.namedWindow('gambar original', cv2.WINDOW_AUTOSIZE)
#     plt.imshow('gambar original', pict)
    cv2.destroyAllWindows()
```
