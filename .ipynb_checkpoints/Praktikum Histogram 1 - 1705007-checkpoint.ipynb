```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img_horse = cv2.imread('download.jpg')
RGB_horse = cv2.cvtColor(img_horse, cv2.COLOR_BGR2RGB)

img_rainbow = cv2.imread('download (1).jfif')
RGB_rainbow = cv2.cvtColor(img_rainbow, cv2.COLOR_BGR2RGB)

img_bricks = cv2.imread('download (2).jfif')
RGB_bricks = cv2.cvtColor(img_bricks, cv2.COLOR_BGR2RGB)
```
