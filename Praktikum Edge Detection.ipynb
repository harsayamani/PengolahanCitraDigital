```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('download.jpg',0)
```

```python
canny = cv2.Canny(img, 100, 200) #canny
sobelx = cv2.Sobel(img, cv2.CV_8U, 1, 0, ksize=3)#sobel
sobely = cv2.Sobel(img, cv2.CV_8U,0,1, ksize=3)
lap = cv2.Laplacian(img, cv2.CV_8U) #laplacian

#prewit
kernelx = np.array([[1,1,1], [0,0,0], [-1,-1,-1]])
prwx = cv2.filter2D(img, -1, kernelx)
kernely = np.array([[1,0,-1], [1,0,-1], [1,0,-1]])
prwy = cv2.filter2D(img, -1, kernely)
```

```python
## tampilkan
plt.subplot(2,3,1), plt.imshow(canny, cmap='gray')
plt.title('canny'), plt.xticks([]), plt.yticks([])

plt.subplot(2,3,2), plt.imshow(sobelx, cmap='gray')
plt.title('sobelx'), plt.xticks([]), plt.yticks([])

plt.subplot(2,3,3), plt.imshow(sobely, cmap='gray')
plt.title('sobely'), plt.xticks([]), plt.yticks([])

plt.subplot(2,3,4), plt.imshow(lap, cmap='gray')
plt.title('laplacian'), plt.xticks([]), plt.yticks([])

plt.subplot(2,3,5), plt.imshow(prwx, cmap='gray')
plt.title('prewitx'), plt.xticks([]), plt.yticks([])

plt.subplot(2,3,6), plt.imshow(prwy, cmap='gray')
plt.title('prewity'), plt.xticks([]), plt.yticks([])
```
